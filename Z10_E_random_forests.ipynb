```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from  mlxtend.plotting  import plot_decision_regions
```

# Random Forests

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"]).astype(np.int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.show()
```

```python
plt.hist(y_train)
plt.hist(y_test)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from  mlxtend.plotting  import plot_decision_regions
```

```python
clf1 = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)
clf1.fit(X_train, y_train)

plot_decision_regions(X_test, y_test, clf1)

```

```python
clf2 = RandomForestClassifier(n_estimators=50, max_leaf_nodes=16, n_jobs=-1, random_state=42)
clf2.fit(X_train, y_train)

plot_decision_regions(X_test, y_test, clf2)

```

```python
clf3 = RandomForestClassifier(n_estimators=5, max_leaf_nodes=16, n_jobs=-1, random_state=42)
clf3.fit(X_train, y_train)

plot_decision_regions(X_test, y_test, clf3)
```
